```python
# Define matrices A, B, C using Python lists
A = [[3.7827, 3.3454, 3.2341], [2.2122, 3.5678, 3.9087], [1.1234, 2.8934, 5.9087]]
B = [[3.1234, 3.0987, 3.1234], [2.1111, 3.2222, 3.3333], [1.0987, 1.3456, 5.1234]]
C = [[3.1243, 3.0989, 3.1256], [2.6721, 3.6785, 3.9017], [1.1254, 2.8956, 5.9187]]

# Function to add two matrices
def add_matrices(matrix1, matrix2):
    result = [[matrix1[i][j] + matrix2[i][j] for j in range(len(matrix1[0]))] for i in range(len(matrix1))]
    return result

# Add matrices A and B
result_add_AB = add_matrices(A, B)

# Print the result of addition
print("Matrix A + Matrix B:")
for row in result_add_AB:
    print(row)

# Example: Add matrices A, B, and C
result_add_ABC = add_matrices(result_add_AB, C)
print("\nMatrix A + Matrix B + Matrix C:")
for row in result_add_ABC:
    print(row)
```

```python
2)	Write a Python Program and perform the following operations ?
I.	Create a List {1225, 4986, 6789, 7890, 2345, 6783, 0987, 1234, 8765, 3456}
II.	Iterate using a for loop
III.	Iterate using for loop and range
IV.	List Comprehension
V.	Enumerate
VI.	Iter function and next function
VII.	Map function
VIII.	Using zip
IX.	Using NumPy Module
```

```python
# Create the list
numbers = [1225, 4986, 6789, 7890, 2345, 6783, 987, 1234, 8765, 3456]
```

```python
# Iterate through the list using a simple for loop
print("Iteration using for loop:")
for num in numbers:
    print(num)
```

```python
# Iterate through the list using a for loop and range
print("\nIteration using for loop and range:")
for i in range(len(numbers)):
    print(numbers[i])
```

```python
# Using list comprehension to create a new list of squared numbers
squared_numbers = [num ** 2 for num in numbers]
print("\nList comprehension (squared numbers):")
print(squared_numbers)
```

```python
# Using enumerate to iterate through the list with an index
print("\nUsing enumerate:")
for index, num in enumerate(numbers):
    print(f"Index {index}: {num}")
```

```python
# Using iter() and next() to manually iterate through the list
print("\nUsing iter() and next():")
numbers_iter = iter(numbers)
try:
    while True:
        print(next(numbers_iter))
except StopIteration:
    pass  # When iteration finishes
```

```python
# Using map() to create a new list of numbers multiplied by 2
multiplied_numbers = list(map(lambda x: x * 2, numbers))
print("\nUsing map function (numbers * 2):")
print(multiplied_numbers)
```

```python
# Using zip to combine two lists element-wise
numbers2 = [i for i in range(10, 20)]  # Another list to zip with
print("\nUsing zip (numbers with another list):")
for num1, num2 in zip(numbers, numbers2):
    print(f"Original: {num1}, Zipped: {num2}")
```

```python
import numpy as np

# Convert the list to a NumPy array and perform operations
numbers_np = np.array(numbers)

# Example operation: add 100 to each element
numbers_np_plus_100 = numbers_np + 100
print("\nUsing NumPy (numbers + 100):")
print(numbers_np_plus_100)
```

```python
3)	For a List A, B, C, D, E write a python program to compute all the combinations and permutations 
```

```python
import itertools

# Define the lists A, B, C, D, E
A = ['a1', 'a2', 'a3']
B = ['b1', 'b2', 'b3']
C = ['c1', 'c2', 'c3']
D = ['d1', 'd2', 'd3']
E = ['e1', 'e2', 'e3']

# Combine all lists into one 
combined_list = A + B + C + D + E

# 1. Compute all combinations of 3 elements from the combined list
combinations = list(itertools.combinations(combined_list, 3))
print("All combinations of 3 elements:")
for comb in combinations:
    print(comb)

# 2. Compute all permutations of 3 elements from the combined list
permutations = list(itertools.permutations(combined_list, 3))
print("\nAll permutations of 3 elements:")
for perm in permutations:
    print(perm)


```

```python
import itertools

# Define the lists A, B, C, D, E
A = ['a1', 'a2', 'a3']
B = ['b1', 'b2', 'b3']
C = ['c1', 'c2', 'c3']
D = ['d1', 'd2', 'd3']
E = ['e1', 'e2', 'e3']

# Combine all lists into one
combined_list = A + B + C + D + E

# 1. Compute all combinations of 3 elements from the combined list
combinations = list(itertools.combinations(combined_list, 3))
print("All combinations of 3 elements:")
for comb in combinations:
    print(comb)

# 2. Compute all permutations of 3 elements from the combined list
permutations = list(itertools.permutations(combined_list, 3))
print("\nAll permutations of 3 elements:")
for perm in permutations:
    print(perm)
```
